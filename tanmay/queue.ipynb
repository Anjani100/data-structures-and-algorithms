```python
class Queue():
    def __init__(self):
        self.rear = None
        self.front = None
    
    def enqueue(self,data,queue):
        node = Node(data)
        node.next = queue.rear
        queue.rear = node
        if queue.front == None:
            queue.front = queue.rear
        return queue
    
    def dequeue(self,queue):
        if queue.rear == None:
            print("Rear empty")
            return queue
        temp = queue.rear
        
        if queue.rear.next == None:
            queue.rear = None
            queue.front = None
            return queue
        
        while(temp.next.next != None):
            print(temp.data,end=" ")
            temp = temp.next
        temp.next = None
        queue.front = temp
        return queue
    
    def create(self,queue):
        queue.rear = None
        queue.front = None
        while True:
            data = int(input("Enter data: (-1) for termination: "))
            if data == -1:
                break
            queue.enqueue(data,queue)
        return queue
    
    def push(self,data,queue):
        queue = queue.enqueue(data,queue)
        return queue
    
    def peak(self,queue):
        return queue.front.data
    
    def pop(self,queue):
        if queue.rear == None:
            return queue
        temp_queue = Queue()
        while(queue.length(queue) != 1):
            data = queue.peak(queue)
            temp_queue.enqueue(data,temp_queue)
            queue.dequeue(queue)
        return temp_queue
        
    def maxElement(self,queue):
        if queue.rear == None:
            return None
        min_data = queue.rear.data
        temp = queue.rear
        while(temp):
            if min_data<temp.data:
                min_data = temp.data
            temp = temp.next
        return min_data
    
    def delete(self,data,queue):
        if queue.rear == None:
            return queue
        if queue.rear.data == data:
            queue.rear = queue.rear.next
            return queue
        temp = queue.rear
        while(temp.next.data != data):
            temp = temp.next
            if temp.next == None:
                return queue

        temp.next = temp.next.next
        return queue
    
    def reverse(self,queue):
        temp_queue = Queue()
        while(queue.length(queue)>1):
            data = queue.maxElement(queue)
            temp_queue.enqueue(data,temp_queue)
            queue.delete(data,queue)
        return temp_queue
    
    def length(self,queue):
        l = 0
        temp = queue.rear
        while(temp):
            l+=1
            temp = temp.next
        return l
    
    def display(slef,queue):
        temp = queue.rear
        while(temp):
            print(temp.data,end=" ")
            temp = temp.next
        print()
```
