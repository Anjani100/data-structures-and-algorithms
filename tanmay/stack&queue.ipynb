```python
class Node():
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class Stack():
    def __init__(self):
        self.head = None
        self.tail = None
    
    def push(self,data):
        newNode = Node(data)
        if self.tail == None:
            self.tail = newNode
            self.head = newNode
            return
        newNode.next = self.head
        self.head = newNode
    
    def pop(self):
        if self.head == None:
            print("Empty")
            return
        self.head = self.head.next
    
    def peek(self):
        if self.head == None:
            print("Empty")
            return
        print(self.head.data)
    
    def isEmpty(self):
        if self.head == None:
            print("Is Empty")
            return
        print("No Empty")
    
    def length(self):
        length = 0
        if self.head == None:
            print(length)
            return
        
        temp = self.head
        while(temp != None):
            length+=1
            temp = temp.next
        print(length)
    
    def display(self):
        if self.head == None:
            print("Empty")
            return
        temp = self.head
        while(temp != None):
            print(temp.data,end = " -> ")
            temp = temp.next
        print("END")
    
```

```python
class Queue():
    def __init__(self):
        self.front = None
        self.rear = None
    
    def enQueue(self,data):
        newQueue = Node(data)
        if self.rear == None:
            self.rear = newQueue
            self.front = newQueue
            return
        newQueue.next = self.rear
        self.rear = newQueue
    
    def deQueue(self):
        if self.rear == None:
            print("Empty")
            return
        
        if self.rear.next == None:
            self.rear = None
            self.front = None
            return
        
        temp = self.rear
        while(temp.next.next):
            temp = temp.next
        
        self.front = temp
        temp.next = None
        
    def length(self):
        length = 0
        if self.rear == None:
            print(length)
            return
        
        temp = self.rear
        while(temp != None):
            length+=1
            temp = temp.next
        print(length)
    
    def display(self):
        if self.rear == None:
            print("Empty")
            return
        
        temp = self.rear
        while(temp != None):
            print(temp.data,end=" -> ")
            temp = temp.next
        print("END")
```

```python
# Queue
if __name__=='__main__':
    queue = Queue()
    while True:
        print('1. Enqueue in queue.')
        print('2. Dequeue in queue.')
        print('3. Length in queue.')
        print('4. Display a queue.')
        ch = int(input('Enter your choice: '))
        if ch == 1:
            data = int(input('Data: '))
            queue.enQueue(data)
        elif ch == 2:
            queue.deQueue()
        elif ch == 3:
            queue.length()
        elif ch == 4:
            queue.display()
        else:
            break
```
