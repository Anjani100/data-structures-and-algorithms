```python
class Node():
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
def isOperand(ch):
    if ch.isalpha() or ch.isnumeric():
        return False
    return True
```

```python
class Stack():
    def __init__(self):
        self.head = None
        self.tail = None
    
    def push(self,data):
        newNode = Node(data)
        if self.tail == None:
            self.tail = newNode
            self.head = newNode
            return
        newNode.next = self.head
        self.head = newNode
    
    def pop(self):
        if self.head == None:
            print("Empty")
            return
        data = self.head.data
        self.head = self.head.next
        return data
    
    def peek(self):
        if self.head == None:
            print("Empty")
            return
        return self.head.data
    
    def isEmpty(self):
        if self.head == None:
            return 1
        return 0
    
    def length(self):
        length = 0
        if self.head == None:
            print(length)
            return
        
        temp = self.head
        while(temp != None):
            length+=1
            temp = temp.next
        print(length)
        
    def makeEmpty(self):
        self.head = None
        self.tail = None
    
    def insertAtBottom(self,data):
        if self.isEmpty():
            self.push(data)
            return
        temp = self.pop()
        self.insertAtBottom(data)
        self.push(temp)
    
    def reverse(self):
        if not self.isEmpty():
            temp = self.pop()
            self.reverse()
            self.insertAtBottom(temp)
    
    def deleteElementAt(self,n):
        count = 0
        if self.head == None:
            return
        temp = self.head
        while count != n:
            if temp == None:
                return
            temp = temp.next
            count+=1
        temp.next = temp.next.next
    
    def display(self):
        if self.head == None:
            print("Empty")
            return
        temp = self.head
        while(temp != None):
            print(temp.data,end = " ")
            temp = temp.next
        print("END")
    
```

```python
class Queue():
    def __init__(self):
        self.front = None
        self.rear = None
    
    def enQueue(self,data):
        newQueue = Node(data)
        if self.rear == None:
            self.rear = newQueue
            self.front = newQueue
            return
        newQueue.next = self.rear
        self.rear = newQueue
    
    def deQueue(self):
        if self.rear == None:
            print("Empty")
            return
        
        if self.rear.next == None:
            self.rear = None
            self.front = None
            return
        
        temp = self.rear
        while(temp.next.next):
            temp = temp.next
        
        self.front = temp
        temp.next = None
        
    def length(self):
        length = 0
        if self.rear == None:
            print(length)
            return
        
        temp = self.rear
        while(temp != None):
            length+=1
            temp = temp.next
        print(length)
    
    def display(self):
        if self.rear == None:
            print("Empty")
            return
        
        temp = self.rear
        while(temp != None):
            print(temp.data,end=" -> ")
            temp = temp.next
        print("END")
```

```python
def towerOfHanoi(n,source,destination,middleware):
    if n == 1:
        print("Move disk ",n," form ",source," to ",destination)
        return
    towerOfHanoi(n-1,source,middleware,destination)
    print("Move disk ", n ," from ", source , " to ", destination)
    towerOfHanoi(n-1,middleware,destination,source)

n = 5
towerOfHanoi(n,'A','B','C')
```
