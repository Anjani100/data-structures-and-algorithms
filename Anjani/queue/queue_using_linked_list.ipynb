```python
class Node():
    def __init__(self, val):
        self.data = val
        self.next = None

class Queue():
    def __init__(self):
        self.head = None
        self.tail = None
    
    def isEmpty(self, queue):
        if not queue.head:
            return True
        else:
            return False
    
    def create_queue(self, queue):
        if not queue.isEmpty(queue):
            print("\n\nQueue is already created!\n\n")
            return queue
        while True:
            print("Enter -1 to exit.")
            n = int(input("Enter the number to be added in the queue: "))
            if n == -1:
                break
            link_node = Node(n)
            if queue.isEmpty(queue):
                queue.head = link_node
                queue.tail = link_node
            else:
                link_node.next = queue.head
                queue.head = link_node
        return queue
    
    def display_queue(self, queue):
        if queue.isEmpty(queue):
            print("\n\nQueue is empty! Cannot display empty queue...\n\n")
            return queue
        disp_node = queue.head
        while disp_node is not None:
            print(disp_node.data, end = " -> ")
            disp_node = disp_node.next
        print("X")
    
    def enqueue(self, queue):
        if queue.isEmpty(queue):
            print("\n\nQueue is empty! Cannot enqueue in an empty queue...\n\n")
            return queue
        n = int(input("Enter the number you want to insert in the queue: "))
        link_node = Node(n)
        link_node.next = queue.head
        queue.head = link_node
        return queue
    
    def dequeue(self, queue):
        if queue.isEmpty(queue):
            print("\n\nQueue is empty! Cannot dequeue in an empty queue...\n\n")
            return queue
        link_node = queue.head
        remove_node = queue.tail
        while link_node.next.next is not None:
            link_node = link_node.next
        link_node.next = None
        remove_node = None
        return queue


if __name__=='__main__':
    queue = Queue()
    while True:
        print('Enter -1 to exit')
        print('1. Create a queue.')
        print('2. Display a queue.')
        print('3. Enqueue.')
        print('4. Dequeue.')
        ch = int(input('Enter your choice: '))
        if ch == 1:
            queue = queue.create_queue(queue)
        elif ch == 2:
            queue.display_queue(queue)
        elif ch == 3:
            queue = queue.enqueue(queue)
        elif ch == 4:
            queue = queue.dequeue(queue)
        else:
            break
```
