```python
class Node():
    def __init__(self, val):
        self.prev = None
        self.data = val
        self.next = None
    
class Deque():
    def __init__(self):
        self.head = None
        self.tail = None
    
    def isEmpty(self, deque):
        if deque.head:
            return False
        return True
    
    def create_deque(self, deque):
        if not deque.isEmpty(deque):
            print('\n\nDeque already created! Use display function to see its content.\n')
            return deque
        while True:
            print('Enter -1 to exit.')
            n = int(input('Enter the number you want to insert: '))
            if n == -1:
                break
            node_link = Node(n)
            if deque.isEmpty(deque):
                deque.head = node_link
                deque.tail = node_link
            else:
                node_link.prev = deque.tail
                deque.tail.next = node_link
                deque.tail = node_link
        return deque
    
    def display_deque(self, deque):
        if deque.isEmpty(deque):
            print('\n\nDeque is empty, cannot display empty deque!\n')
            return
        disp_node = deque.head
        while disp_node is not None:
            print(disp_node.data, end = " -> ")
            disp_node = disp_node.next
        print('X')
    
    def insertFront(self, deque):
        n = int(input('Enter the number you want to insert: '))
        node_link = Node(n)
        if deque.isEmpty(deque):
            deque.head = node_link
            deque.tail = node_link
        else:
            node_link.next = deque.head
            deque.head.prev = node_link
            deque.head = node_link
        return deque
    
    def insertEnd(self, deque):
        n = int(input('Enter the number you want to insert: '))
        node_link = Node(n)
        if deque.isEmpty(deque):
            deque.head = node_link
            deque.tail = node_link
        else:
            node_link.prev = deque.tail
            deque.tail.next = node_link
            deque.tail = node_link
        return deque
    
    def deleteFront(self, deque):
        if deque.isEmpty(deque):
            print('\n\nDeque is empty, cannot delete from empty deque!\n')
            return deque
        if deque.head.next is None:
            deque.head = None
            deque.tail = None
            return deque
        remove_node = deque.head
        deque.head = deque.head.next
        remove_node = None
        return deque
    
    def deleteEnd(self, deque):
        if deque.isEmpty(deque):
            print('\n\nDeque is empty, cannot delete from empty deque!\n')
            return deque
        if deque.head.next is None:
            deque.head = None
            deque.tail = None
            return deque
        remove_node = deque.tail
        deque.tail = deque.tail.prev
        deque.head.prev = None
        remove_node = None
        return deque


if __name__=='__main__':
    deque = Deque()
    while True:
        print('Enter -1 to exit.')
        print('1. Create a deque.')
        print('2. Display the deque.')
        print('3. Insert at Front.')
        print('4. Insert at End.')
        print('5. Delete from Front.')
        print('6. Delete from End.')
        ch = int(input('Enter your choice: '))
        if ch == 1:
            deque = deque.create_deque(deque)
        elif ch == 2:
            deque.display_deque(deque)
        elif ch == 3:
            deque = deque.insertFront(deque)
        elif ch == 4:
            deque = deque.insertEnd(deque)
        elif ch == 5:
            deque = deque.deleteFront(deque)
        elif ch == 6:
            deque = deque.deleteEnd(deque)
        else:
            break
```
