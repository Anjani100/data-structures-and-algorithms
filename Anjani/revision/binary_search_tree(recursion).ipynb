```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def insert(root, val):
    if not root:
        return Node(val)
    if root.val == val:
        return root
    if root.val > val:
        root.left = insert(root.left, val)
    else:
        root.right = insert(root.right, val)
    return root

def preorder(root):
    if root:
        print(root.val, end = ' ')
        preorder(root.left)
        preorder(root.right)

def inorder(root):
    if root:
        inorder(root.left)
        print(root.val, end = ' ')
        inorder(root.right)

def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.val, end = ' ')

def search(root, val):
    if not root:
        return False
    if root.val == val:
        return True
    if root.val > val:
        return search(root.left, val)
    else:
        return search(root.right, val)

def get_max(root):
    while root.right:
        root = root.right
    return root

def get_min(root):
    while root.left:
        root = root.left
    return root

def delete(root, val):
    if not search(root, val):
        return False
    if not root:
        return root
    if root.val > val:
        root.left = delete(root.left, val)
    elif root.val < val:
        root.right = delete(root.right, val)
    else:
        if not root.left:
            temp = root.right
            root = None
            return temp
        if not root.right:
            temp = root.left
            root = None
            return temp
        temp = get_min(root.right)
        root.val = temp.val
        root.right = delete(root.right, temp.val)
    return root

def check_bst(root, smallest = float('-inf'), largest = float('inf')):
    if not root:
        return True
    if root.val < smallest or root.val > largest:
        return False
    return check_bst(root.left, smallest, root.val - 1) and check_bst(root.right, root.val + 1, largest)

if __name__=='__main__':
    root = None
    while True:
        print('\nEnter -1 to exit.')
        print('1. Insert a value.')
        print('2. Preorder.')
        print('3. Inorder.')
        print('4. Postorder.')
        print('5. Search for a value.')
        print('6. Get Max value.')
        print('7. Get Min value.')
        print('8. Delete a value.')
        print('9. Check if a given tree is a BST.')
        ch = int(input('Enter your choice: '))
        if ch == 1:
            while True:
                print('Enter -1 to exit.')
                num = int(input('Enter the number you want to insert: '))
                if num == -1:
                    break
                root = insert(root, num)
        elif ch == 2:
            preorder(root)
        elif ch == 3:
            inorder(root)
        elif ch == 4:
            postorder(root)
        elif ch == 5:
            num = int(input('Enter the number you want to search: '))
            if search(root, num):
                print('Found!')
            else:
                print('Not Found!')
        elif ch == 6:
            print(get_max(root).val)
        elif ch == 7:
            print(get_min(root).val)
        elif ch == 8:
            num = int(input('Enter the number you want to delete: '))
            root = delete(root, num)
        elif ch == 9:
            root = Node(4)
            root.left = Node(2)
            root.right = Node(5)
            root.left.left = Node(1)
            root.left.right = Node(3)
            if(check_bst(root)):
                print('The given tree is a BST.')
            else:
                print('The given tree is not a BST.')
        elif ch == -1:
            break
        else:
            print('Wrong choice! Please choose a valid option.')
```
