```python
class Node():
    def __init__(self, val):
        self.data = val
        self.next = None
    
class Stack():
    def __init__(self):
        self.head = None
        self.tail = None
    
    def create_stack(self, stack):
        if stack.head:
            print('\n\nStack is already created!\n')
            return stack
        while True:
            print('\n\nEnter -1 to exit.')
            n = int(input('Enter the number you want to insert: '))
            if n == -1:
                break
            push_link = Node(n)
            if not stack.head:
                stack.head = push_link
                stack.tail = push_link
            else:
                push_link.next = stack.head
                stack.head = push_link
        return stack
    
    def display_stack(self, stack):
        if not stack.head:
            print('Stack is empty. Cannot display empty stack!')
            return stack
        disp_link = stack.head
        while disp_link is not None:
            print('|', disp_link.data, '|', end = "\n")
            disp_link = disp_link.next
        print('  X')
    
    def push(self, stack, n):
        push_link = Node(n)
        push_link.next = stack.head
        stack.head = push_link
        return stack
    
    def pop(self, stack):
        if not stack.head:
#             print('Stack is empty. Cannot pop in empty stack!')
            return -1, stack
        popped_element = stack.head.data
        pop_link = stack.head
        stack.head = stack.head.next
        pop_link = None
        return popped_element, stack
    
    def peek(self, stack):
        if not stack.head:
            print('Stack is empty. Cannot peek in empty stack!')
            return stack
        return stack.head.data
    
    def isEmpty(self, stack):
        if not stack.head:
            return True
        else:
            return False
    
    def evaluate_postfix(self, stack, expression):
        expression = expression.split()
        for i in expression:
            if not i.isdigit():
                first_operand, stack = stack.pop(stack)
                second_operand, stack = stack.pop(stack)
                if i == '+':
                    result = second_operand + first_operand
                elif i == '-':
                    result = second_operand - first_operand
                elif i == '*':
                    result = second_operand * first_operand
                elif i == '/':
                    result = second_operand / first_operand
                stack = stack.push(stack, result)
            else:
                stack = stack.push(stack, int(i))
        print('\n\nThe evaluation of the given postfix is', stack.peek(stack))
        return stack
    
    def evaluate_prefix(self, stack, expression):
        expression = expression.split()
        expression.reverse()
        for i in expression:
            if not i.isdigit():
                first_operand, stack = stack.pop(stack)
                second_operand, stack = stack.pop(stack)
                if i == '+':
                    result = first_operand + second_operand
                elif i == '-':
                    result = first_operand - second_operand
                elif i == '*':
                    result = first_operand * second_operand
                elif i == '/':
                    result = first_operand / second_operand
                stack = stack.push(stack, result)
            else:
                stack = stack.push(stack, int(i))
        print('\n\nThe evaluation of the given prefix is', stack.peek(stack))
        return stack
    
    def hasHigherPrecedence(self, first_operand, second_operand):
        if first_operand == '+' or first_operand == '-':
            first_weight = 1
        elif first_operand == '*' or first_operand == '/':
            first_weight = 2
        if second_operand == '+' or second_operand == '-':
            second_weight = 1
        elif second_operand == '*' or second_operand == '/':
            second_weight = 2
        if first_weight <= second_weight:
            return True
        else:
            return False
    
    def isOpeningParenthesis(self, operator):
        if operator == "(" or operator == "{" or operator == "[":
            return True
        else:
            return False
    
    def isClosingParenthesis(self, operator):
        if operator == ")" or operator == "}" or operator == "]":
            return True
        else:
            return False
    
    def infix_to_postfix(self, stack, expression):
        expression = expression.split()
        postfix_expression = ""
        for i in expression:
            if i.isalnum():
                postfix_expression += i + " "
            else:
                while not stack.isEmpty(stack) and stack.hasHigherPrecedence(i, stack.peek(stack)):
                    popped_element, stack = stack.pop(stack)
                    postfix_expression += popped_element + " "
                stack = stack.push(stack, i)
        while not stack.isEmpty(stack):
            popped_element, stack = stack.pop(stack)
            postfix_expression += popped_element + " "
        print('\n\nThe equivalent Postfix expression to the given Infix expression is', postfix_expression)
        return stack
    
    def infix_to_postfix_brackets_handled(self, stack, expression):
        expression = expression.split()
        postfix_expression = ""
        for i in expression:
            if i.isalnum():
                postfix_expression += i + " "
            elif stack.isOpeningParenthesis(i):
                stack = stack.push(stack, i)
            elif stack.isClosingParenthesis(i):
                while not stack.isEmpty(stack) and not stack.isOpeningParenthesis(stack.peek(stack)):
                    popped_element, stack = stack.pop(stack)
                    postfix_expression += popped_element + " "
                popped_element, stack = stack.pop(stack)
            else:
                while not stack.isEmpty(stack) and not stack.isOpeningParenthesis(stack.peek(stack)) and stack.hasHigherPrecedence(i, stack.peek(stack)):
                    popped_element, stack = stack.pop(stack)
                    postfix_expression += popped_element + " "
                stack = stack.push(stack, i)
        while not stack.isEmpty(stack):
            popped_element, stack = stack.pop(stack)
            postfix_expression += popped_element + " "
        print('\n\nThe equivalent Postfix expression to the given Infix expression is', postfix_expression)
        return stack
    
    def infix_to_prefix(self, stack, expression):
        expression = expression.split()
        expression.reverse()
        prefix_expression = ""
        for i in expression:
            if i.isalnum():
                prefix_expression += i + " "
            else:
                while not stack.isEmpty(stack) and stack.hasHigherPrecedence(i, stack.peek(stack)):
                    popped_element, stack = stack.pop(stack)
                    prefix_expression += popped_element + " "
                stack = stack.push(stack, i)
        while not stack.isEmpty(stack):
            popped_element, stack = stack.pop(stack)
            prefix_expression += popped_element + " "
        print('\n\nThe equivalent Prefix expression to the given Infix expression is', prefix_expression[::-1])
        return stack
    
    def infix_to_prefix_brackets_handled(self, stack, expression):
        expression = expression.split()
        expression.reverse()
        prefix_expression = ""
        for i in expression:
            if i.isalnum():
                prefix_expression += i + " "
            elif stack.isClosingParenthesis(i):
                stack = stack.push(stack, i)
            elif stack.isOpeningParenthesis(i):
                while not stack.isEmpty(stack) and not stack.isClosingParenthesis(stack.peek(stack)):
                    popped_element, stack = stack.pop(stack)
                    prefix_expression += popped_element + " "
                _, stack = stack.pop(stack)
            else:
                while not stack.isEmpty(stack) and not stack.isClosingParenthesis(stack.peek(stack)) and stack.hasHigherPrecedence(i, stack.peek(stack)):
                    popped_element, stack = stack.pop(stack)
                    prefix_expression += popped_element + " "
                stack = stack.push(stack, i)
        while not stack.isEmpty(stack):
            popped_element, stack = stack.pop(stack)
            prefix_expression += popped_element + " "
        print('\n\nThe equivalent Prefix expression to the given Infix expression is', prefix_expression[::-1])
        return stack
    
    def postfix_to_infix(self, stack, expression):
        expression = expression.split()
        for i in expression:
            if i.isalnum():
                stack = stack.push(stack, i)
            else:
                second_operand, stack = stack.pop(stack)
                first_operand, stack = stack.pop(stack)
                result = "( " + first_operand + " " + i + " " + second_operand + " )"
                stack = stack.push(stack, result)
        print('\n\nThe equivalent Infix expression to the given Postfix expression is', stack.peek(stack))
        return stack
    
    def postfix_to_prefix(self, stack, expression):
        expression = expression.split()
        for i in expression:
            if i.isalnum():
                stack = stack.push(stack, i)
            else:
                second_operand, stack = stack.pop(stack)
                first_operand, stack = stack.pop(stack)
                result = i + " " + first_operand + " " + second_operand
                stack = stack.push(stack, result)
        print('\n\nThe equivalent Prefix expression to the given Postfix expression is', stack.peek(stack))
        return stack
    
    def prefix_to_infix(self, stack, expression):
        expression = expression.split()
        expression.reverse()
        for i in expression:
            if i.isalnum():
                stack = stack.push(stack, i)
            else:
                first_operand, stack = stack.pop(stack)
                second_operand, stack = stack.pop(stack)
                result = first_operand + " " + i + " " + second_operand
                stack = stack.push(stack, result)
        print('\n\nThe equivalent Infix expression to the given Prefix expression is', stack.peek(stack))
        return stack
    
    def prefix_to_postfix(self, stack, expression):
        expression = expression.split()
        expression.reverse()
        for i in expression:
            if i.isalnum():
                stack = stack.push(stack, i)
            else:
                first_operand, stack = stack.pop(stack)
                second_operand, stack = stack.pop(stack)
                result = first_operand + " " + second_operand + " " + i
                stack = stack.push(stack, result)
        print('\n\nThe equivalent Postfix expression to the given Prefix expression is', stack.peek(stack))
        return stack
    
    def display_toh_moves(self, s, d, disk):
        print('Move disk', disk, 'from', s, 'to', d)
    
    def move_disks(self, source, dest, s, d):
        first_popped_element, source = source.pop(source)
        second_popped_element, dest = dest.pop(dest)
        if first_popped_element == -1:
            source = source.push(source, second_popped_element)
            source.display_toh_moves(d, s, second_popped_element)
        elif second_popped_element == -1:
            dest = dest.push(dest, first_popped_element)
            dest.display_toh_moves(s, d, first_popped_element)
        elif first_popped_element > second_popped_element:
            source = source.push(source, first_popped_element)
            source = source.push(source, second_popped_element)
            source.display_toh_moves(d, s, second_popped_element)
        else:
            dest = dest.push(dest, second_popped_element)
            dest = dest.push(dest, first_popped_element)
            dest.display_toh_moves(s, d, first_popped_element)
        
        return source, dest
    
    def tower_of_hanoi(self, no_of_disks, source, aux, dest):
        print('\n\n')
        s = 'S'
        a = 'A'
        d = 'D'
        if no_of_disks % 2 == 0:
            a, d = d, a
        no_of_moves = pow(2, no_of_disks) - 1
        for i in range(no_of_disks, 0, -1):
            source = source.push(source, i)
            
        for i in range(no_of_moves):
            if i % 3 == 0:
                source, dest = source.move_disks(source, dest, s, d)
            elif i % 3 == 1:
                source, aux = source.move_disks(source, aux, s, a)
            elif i % 3 == 2:
                aux, dest = source.move_disks(aux, dest, a, d)
        print('\n\n')
    
    def insertAtBottom(self, stack, data):
        if stack.isEmpty(stack):
            stack = stack.push(stack, data)
            return stack
        popped_element, stack = stack.pop(stack)
        stack = stack.insertAtBottom(stack, data)
        stack = stack.push(stack, popped_element)
        return stack
    
    def reverse_stack(self, stack):
        if not stack.isEmpty(stack):
            popped_element, stack = stack.pop(stack)
            stack = stack.reverse_stack(stack)
            stack = stack.insertAtBottom(stack, popped_element)
        return stack

if __name__=='__main__':
    stack = Stack()
    while True:
        print('Enter -1 to exit.')
        print('1. Create a stack.')
        print('2. Display a stack.')
        print('3. Push an element.')
        print('4. Pop an element.')
        print('5. Postfix Evaluation.')
        print('6. Prefix Evaluation.')
        print('7. Infix to Postfix Evaluation.')
        print('8. Advanced Infix to Postfix Evaluation (With Brackets).')
        print('9. Infix to Prefix Evaluation.')
        print('10. Advanced Infix to Prefix Evaluation (With Brackets).')
        print('11. Postfix to Infix Evaluation.')
        print('12. Postfix to Prefix Evaluation.')
        print('13. Prefix to Infix Evaluation.')
        print('14. Prefix to Postfix Evaluation.')
        print('15. Tower of Hanoi (no recursion)')
        print('16. Reverse a stack using recursion (no iterations)')
        ch = int(input('Enter your choice: '))
        if ch == 1:
            stack = stack.create_stack(stack)
        elif ch == 2:
            stack.display_stack(stack)
        elif ch == 3:
            n = int(input('Enter the number you want to push: '))
            stack = stack.push(stack, n)
        elif ch == 4:
            popped_element, stack = stack.pop(stack)
        elif ch == 5:
            expression = input('Enter the postfix expression: ')
            stack = stack.evaluate_postfix(stack, expression)
        elif ch == 6:
            expression = input('Enter the prefix expression: ')
            stack = stack.evaluate_prefix(stack, expression)
        elif ch == 7:
            expression = input('Enter the infix expression: ')
            stack = stack.infix_to_postfix(stack, expression)
        elif ch == 8:
            expression = input('Enter the infix expression: ')
            stack = stack.infix_to_postfix_brackets_handled(stack, expression)
        elif ch == 9:
            expression = input('Enter the infix expression: ')
            stack = stack.infix_to_prefix(stack, expression)
        elif ch == 10:
            expression = input('Enter the infix expression: ')
            stack = stack.infix_to_prefix_brackets_handled(stack, expression)
        elif ch == 11:
            expression = input('Enter the postfix expression: ')
            stack = stack.postfix_to_infix(stack, expression)
        elif ch == 12:
            expression = input('Enter the postfix expression: ')
            stack = stack.postfix_to_prefix(stack, expression)
        elif ch == 13:
            expression = input('Enter the prefix expression: ')
            stack = stack.prefix_to_infix(stack, expression)
        elif ch == 14:
            expression = input('Enter the prefix expression: ')
            stack = stack.prefix_to_postfix(stack, expression)
        elif ch == 15:
            source = Stack()
            aux = Stack()
            dest = Stack()
            no_of_disks = int(input('Enter the number of disks: '))
            stack.tower_of_hanoi(no_of_disks, source, aux, dest)
        elif ch == 16:
            if stack.isEmpty(stack):
                print('\n\nCannot reverse an empty stack. Create one first!\n\n')
            else:
                stack = stack.reverse_stack(stack)
        else:
            break
```
